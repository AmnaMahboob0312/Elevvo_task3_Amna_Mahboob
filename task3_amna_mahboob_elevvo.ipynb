```python
print("Q4 Education level unique values: ", new_df['Q4_What_is_the_highest_level_of_formal _education_that_you_have_attained_or_plan_to_attain_within_the_next_ 2_ years?'].unique())
```

```python
degree_mapping = {
    'Bachelorâ€™s degree': "Bachelor's degree",
    'Masterâ€™s degree': "Master's degree",
    'Some college/university study without earning a bachelorâ€™s degree': "Some college/university study without earning a bachelor's degree"
}

new_df['Q4_What_is_the_highest_level_of_formal _education_that_you_have_attained_or_plan_to_attain_within_the_next_ 2_ years?'] = new_df['Q4_What_is_the_highest_level_of_formal _education_that_you_have_attained_or_plan_to_attain_within_the_next_ 2_ years?'].replace(degree_mapping, regex=True)

print("Q4 Education level unique values after cleaning: ", new_df['Q4_What_is_the_highest_level_of_formal _education_that_you_have_attained_or_plan_to_attain_within_the_next_ 2_ years?'].unique())
```

step 3 - do categorical handling

```python
new_df.info()
```

```python
col_select_to_one_col = ['Q7_Part_1_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Python','Q7_Part_2_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - R','Q7_Part_3_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - SQL','Q7_Part_4_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - C','Q7_Part_5_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - C++','Q7_Part_6_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Java','Q7_Part_7_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Javascript','Q7_Part_8_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Julia','Q7_Part_9_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Swift','Q7_Part_10_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Bash','Q7_Part_11_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - MATLAB','Q7_Part_12_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - None','Q7_Part_13_What_programming_languages_do_you_use_on_a_regular_basis?_Selected_Choice - Other']
new_df['Q7_Programming_language_used_daily'] = new_df[col_select_to_one_col].apply(lambda x:x.dropna().iloc[0] if not x.dropna().empty else np.nan, axis=1)
new_df['Q7_Programming_language_used_daily'] = new_df['Q7_Programming_language_used_daily'].astype("category")
```

```python
new_df.head(10)
```

```python
remove_cols = col_select_to_one_col
removed_cols = [col for col in remove_cols if new_df[col].isnull().any()]
new_df = new_df.drop(columns=removed_cols)
```

```python
new_df.head(6)
```

```python
print("Q2 gender unique values: ", new_df['Q2_What_is_your_gender? '].unique())
```

```python
gender_mapping = {
    'Male': "Man",
    'Female': "Woman",
    'Nonbinary': "Non-binary, genderqueer, or gender non-conforming",
    'A different identity' : "Prefer not to say"
}

new_df['Q2_What_is_your_gender? '] = new_df['Q2_What_is_your_gender? '].replace(gender_mapping, regex=True)

print("Q2 gender unique values after cleaning: ", new_df['Q2_What_is_your_gender? '].unique())
```
